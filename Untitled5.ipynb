```python
import pandas as pd
file_path_corrected = '/content/movie.xls'

df = pd.read_csv(file_path_corrected)
```

```python
# 1.
missing_values = df['duration'].isnull().sum()
missing_values
```

```python
# 2.
median_duration = df['duration'].median()
df['duration'].fillna(median_duration, inplace=True)
missing_values = df['duration'].isnull().sum()
missing_values
```

```python
# 3.
df['genres_count'] = df['genres'].apply(lambda x: len(x.split('|')))
average_genres = df['genres_count'].mean()
average_genres
```

```python
#4.
df['genre_count'] = df['genres'].apply(lambda x: len(x.split('|')))

max_genres = df['genre_count'].max()

max_genre_movies = df[df['genre_count'] == max_genres]

max_genre_movies[['movie_title', 'genres', 'genre_count']]
```

```python
# 5.
all_genres = df['genres'].str.split('|').explode()
most_common= all_genres.mode()[0]
most_common
```

```python
# 6.
df['budget'] = df['budget'].str.replace('$', '').astype(float)


filtered_df = df[(df['title_year'] >= 2014) & (df['title_year'] <= 2016)]


pivot_table = filtered_df.pivot_table(values='budget', index='country', columns='title_year', aggfunc='sum', fill_value=0)

pivot_table
```
